```python
autos["brand"].value_counts(normalize=True).head(20)
```

The brands that I have chosen to aggregate for are the top 20 brands listed above.

```python
brands_mean = {}
brands = list(autos["brand"].value_counts(normalize=True,dropna=False).head(20).index)
#make the top 20 brands into a list to loop through
```

```python
for brand in brands:
    brands_mean[brand] = autos.loc[autos["brand"] == brand, "price"].mean()
brands_mean
```

Mercedes, sonstige_autos and citroen have the highest average price. While mazda and toyota are lowest.

```python
mileage_mean = {}
for brand in brands:
    mileage_mean[brand] = autos.loc[autos["brand"] == brand, "odometer_km"].mean()
mileage_mean
```

```python
df_mean_price = pd.DataFrame(pd.Series(brands_mean), columns=["mean price"])
df_mean_mileage = pd.DataFrame(pd.Series(mileage_mean), columns=["mean mileage"])
```
