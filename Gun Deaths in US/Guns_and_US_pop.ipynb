```python
import csv

file = open("guns.csv", "r")
data = list(csv.reader(file))
print(data[:5])
```

```python
headers = data[0]
data = data[1:]
print(headers)
print(data[:5])
```

```python
years = [row[1] for row in data]
years_count = {}
for year in years:
    if year in years_count:
        years_count[year] += 1
    else:
        years_count[year] = 1
        
for key, val in years_count.items():
    print(key, val)
```

```python
import datetime
dates = [datetime.datetime(year=int(row[1]),month=int(row[2]), day=1) for row in data]
dates[:5]
```

```python
date_counts = {}
for date in dates:
    if date not in date_counts:
        date_counts[date] = 1
    else:
        date_counts[date] += 1
date_counts
```

```python
sex_counts = {} #5
race_counts = {} #7

for row in data:
    if row[5] not in sex_counts:
        sex_counts[row[5]] = 1
    else:
        sex_counts[row[5]] += 1
        
    if row[7] not in race_counts:
        race_counts[row[7]] = 1
    else:
        race_counts[row[7]] += 1
print(race_counts)
print(sex_counts)
```
