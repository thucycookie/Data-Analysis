```python
Sample_size_vs_median = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
Sample_size_vs_median.set_title('Sample size vs. median income')
```

```python
Sample_size_vs_unemployment_rate = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
Sample_size_vs_unemployment_rate.set_title('Sample size vs. Unemployment Rate')
```

```python
Full_time_vs_median = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
Full_time_vs_median.set_title('Full time vs. Median')
```

```python
ShareWomen_vs_Unemployment_rate = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ShareWomen_vs_Unemployment_rate.set_title('ShareWomen vs. Unemployment Rate')
```

```python
Men_vs_median = recent_grads.plot(x='Men', y='Median', kind='scatter')
Men_vs_median.set_title('Men vs. Median Income')
```

```python
Women_vs_median = recent_grads.plot(x='Women', y='Median', kind='scatter')
Women_vs_median.set_title('Women vs. Median Income')
```

Students in more popular majors do not necessariy make more money because there seems to be no upward trend in the Sample_size vs. Median income graph.

Students that majored in subjects that were majority female do not make more money comparative to male.

There is no strong link between the number of full-time employees and median salary. 

```python
ax = recent_grads["Sample_size"].hist(bins=50,range=(0,5000))
ax.set_xlabel("Sample size")
ax.set_ylabel("Frequency")
ax.set_title("Distribution of the sample size")
```

```python
ax = recent_grads["Median"].hist(bins=50,range=(15000,100000))
ax.set_xlabel("Median")
ax.set_ylabel("Frequency")
ax.set_title("Distribution of the median")
```

```python
ax = recent_grads["Employed"].hist(bins=50,range=(15000,310000))
ax.set_xlabel("Employed")
ax.set_ylabel("Frequency")
ax.set_title("Distribution of the employed")
print(recent_grads["Employed"].describe())
```

```python
ax = recent_grads["Full_time"].hist(bins=50,range=(15000,290000))
ax.set_xlabel("Full time")
ax.set_ylabel("Frequency")
ax.set_title("Distribution of the full time")
print(recent_grads["Full_time"].describe())
```

```python
ax = recent_grads["ShareWomen"].hist(bins=50,range=(0,1))
ax.set_xlabel("ShareWomen")
ax.set_ylabel("Frequency")
ax.set_title("Distribution of ShareWomen")
print(recent_grads["ShareWomen"].describe())
```

```python
ax = recent_grads["Unemployment_rate"].hist(bins=100,range=(0,0.2))
ax.set_xlabel("Unemployment_rate")
ax.set_ylabel("Frequency")
ax.set_title("Distribution of unemployment rate")
print(recent_grads["Unemployment_rate"].describe())
```

```python
ax = recent_grads["Men"].hist(bins=50,range=(0,200000))
ax.set_xlabel("Men")
ax.set_ylabel("Frequency")
ax.set_title("Distribution of male-dominated majors")
print(recent_grads["Men"].describe())
```

```python
ax = recent_grads["Women"].hist(bins=100,range=(0,310000))
ax.set_xlabel("Female-dominated majors")
ax.set_ylabel("Frequency")
ax.set_title("Distribution of female-dominated majors")
print(recent_grads["Women"].describe())
```

The majors that are dominated by men siginficantly greater than the majors that are dominated by women.

The most common median salary range is around 30,000 to 50,000. The peak is around 35,000 dollars
